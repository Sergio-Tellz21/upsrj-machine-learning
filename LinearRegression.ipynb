#Importaci贸n de librerias


```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#Creacion de datos de prueba

```python
#y = m*x+b
#Variable independiente
#ScKit-Learn requiere que se usen arreglos de 2D
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
print(X)

#Varialble dependiente
y=np.array([2, 4, 5, 4, 5])
print(y)
```

#Graficacion de datos originales

```python
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Dtos originales")
plt.grid(True)
plt.show()
```

#Generacion y alimentacion del modelo

```python
#Generamos nuestro modelo de regresion linear
model= LinearRegression()
#alimentamos nuestro modelo

model.fit(X,y)

#parametros:m(peso), b(sesgo)

m=model.coef_[0]
b = model.intercept_

print(f"Pendiente (m):{m:2f}")
print(f"Intercepto(b):{b:2f}")
```

#prediccion y evaluacion del modelo

```python
#generamos una predicci贸n
y_pred = model.predict(X)
#Evaluamos esa predicci贸n
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y,y_pred)

print(f"Mean Squared Error:{mse:2f}")
print(f"R2 Score:{r2:2f}")
```

#graficaci贸n del modelo

```python
plt.scatter(X,y, label="Datos reales")
plt.plot(X,y_pred,color="red", label="Modelo lineal")
plt.xlabel("X")
plt.ylabel("y")
plt.scatter(X,y)
plt.title("Regresion linear simple")
plt.grid(True)
plt.show()
```

# Implementacion de Pipeline

Un pipeline encapsula todo el fluo de procesamiento de datos en un solo objeto.
para el Usuario, entrenar un pipeline es igual que entregar un modelo.
El escalado no cambia el resultado conceptual, pero prepara el terreno para modelos m谩s complejos.

## Justificacion:

Hasta ahora hemos entrenado paso por paso.
En la vida real, esos pasos siempre se repiten en el mismo orden.
Un Pipeline es una forma de decirle a la computadora: **Estos son los pasos, siempre ejecutalos as铆**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
def plot_model(xlabel: str, ylabel: str, title:str):
    plt.scatter(X,y, label="datos reales")
    plt.plot(X, y_pred, color="red",label="modelo lineal")
    plt.grid(visible=True)
    plt.show()
```

```python
#1. definiremos las variables dependientes e independientes de nuestro modelo.

#Variable independiente X: Distancia al centro(km)
X = np.arange(1,21).reshape(-1,1)
print(X)

#Variable dependiente y: Precio(miles de $)
y = np.array([340, 320, 290, 260, 230,
              210, 220, 245, 275, 305,
              330, 350, 365, 375, 380,
              385, 390, 395, 398, 400])
```

```python
#Definimos del pipeline de regresion lineal
pipeline = Pipeline( steps=[
    ("Scaler", StandardScaler()),
    ("regressor", LinearRegression())
])
#Entrenamiento del modelo
pipeline.fit(X,y)

#Prediccion con el modelo
y_pred = pipeline.predict(X)

plot_model(xlabel="Distancia al centro(Km)", ylabel="Precio (miles de $)",
           title ="Estimacion de precio de una casa (regresion lineal simple)")
```
